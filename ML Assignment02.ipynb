## 1. Definition


> A Taylor series approximation uses a Taylor series to represent a number as a polynomial that has a very similar value to the number in a neighborhood around a specified $x$ value :

$$f(x)=f(a) + \frac{f'(a)}{1!}(x - a) + \frac{f''(a)}{2!}(x - a)^2 + \dots + \frac{f^{(n)}(a)}{n!}(x - a)^n + \dots$$

## 2. First-order Talyor Approximation
- First-order Talyor approximation of $f$, near point $z$ : 

$$\hat{f}(x) = f(z) + \frac{\partial f}{\partial x_1}(z)(x_1 - z_1) + \dots + \frac{\partial f}{\partial x_n}(z)(x_n - z_n) $$
- $\hat{f}(x)$ is very close to $f(x)$ when $x_i$ are all near $z_i$
- $\hat{f}$ is an affine function of $x$
- can write using inner product as

$$ \hat{f}(x) = f(z) + \nabla f(z)^T (x - z), \left( \nabla f(z) = \left( \frac{\partial f}{\partial x_1}(z), \dots , \frac{\partial f}{\partial x_n}(z)\right) \right) $$

## 3. Example
- I used vertical displacement function $f(t)$ for parabolic motion for time $t \in [0, 20]$

 - $f(t) = v_0 t - \frac{1}{2} gt^2, (v_0 = 100, g = 9.8)$

 - $domain$   $t \in [0,20] $
 
### (1) Origin function $f(t)$ has been plotted like below.


```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np

v0 = 100
g = 9.8
t = np.linspace(0,20)
s = v0 * t - 1/2 * g * np.power(t, 2)

plt.grid(True)
plt.xlabel('t')
plt.ylabel('s')
plt.title('Parabolic Motion_Vertical Element')

plt.plot(t, s, label = r'$f(t) = v_0 t - \frac{1}{2} gt^2$')
plt.legend()
plt.show()
```

### (2) for $t=15$, $f(t)$ go through the point $(15, 397.5)$

```python
v0 = 100
g = 9.8
t = np.linspace(0,20)
s = v0 * t - 1/2 * g * np.power(t, 2)
plt.grid(True)
plt.xlabel('t')
plt.ylabel('s')
plt.title('Parabolic Motion_Vertical Element')

plt.plot(t, s, label = r'$f(t) = v_0 t - \frac{1}{2} gt^2$')
plt.plot(15,397.5,'ro')
plt.legend()
plt.show()
```

### (3) Now we can get a linear function $\hat{f}(t)$ For the point $(15,397.5)$ using first-order Talyor approximation.
$\hat{f}(t) = f(15) + \nabla f(15)^T (t - 15)$

$\Rightarrow \nabla f(t) = \left( \frac{\partial f}{\partial t}f(t) \right) $

$\Rightarrow t = 15, \nabla f(15) = f'(15) = -47$

$\Rightarrow \hat{f}(t) = f(15) + \nabla f(15)^T (t - 15) $

$\Leftrightarrow \hat{f}(t) = 397.5 - 47(t - 15) $

$\Leftrightarrow \hat{f}(t) = 147t - 1807.5$

### (4) $f(t)$ and $\hat{f}(t)$ has been plotted like below at the point $(15, 397.5)$

```python
v0 = 100
g = 9.8
t = np.linspace(0,20)
s = v0 * t - 1/2 * g * np.power(t, 2)
l = -47 * t + 1102.5

plt.grid(True)
plt.xlabel('t')
plt.ylabel('s')
plt.title('Parabolic Motion_Vertical Element')

plt.plot(t, s, label = r'$f(t) = v_0 t - \frac{1}{2} gt^2$')
plt.plot(t, l, label = '$\hat{f}(t) = -47t + 1102.5$')
plt.plot(15,397.5,'ro')

plt.legend()
plt.show()
```
